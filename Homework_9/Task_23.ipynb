```python
import pandas as pd
import sqlite3
with sqlite3.connect('population.db') as con:
    df = pd.read_sql('select * from population', con=con)
```

```python
df_male_in_Califonia = df[(df['gender'] == 'Male') & (df['state'] == 'California')]
df_mean = df_male_in_Califonia['salary'].mean()
df = df[df['salary'] < df_mean]
df
```
